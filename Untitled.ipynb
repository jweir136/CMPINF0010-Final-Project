```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Pitt Pools.csv', index_col='id')
data.drop(columns=['retired', 'council_district', 'ward', 'tract'])
```

```python
pool_type = data['type'].value_counts().sort_values(ascending=False)
pool_type
```

```python
pool_type.plot.pie(title='Total # of Pools and Spray Parks in Pittsburgh')
```

```python
community_pools = data['neighborhood'].value_counts().sort_values(ascending=False)
data['neighborhood'].value_counts()
```

```python
community_pools.plot.bar(title='# of Pools & Spray Parks per Neighborhood')
```

## Conclusion

From this data we can see that South Side Flats and Highland Park have the most amount of pool ameneties wiithin their neighborhood. They both have two pools respectively, while they neither have a spray park one would assume having and maintaining a pool is a higher cost. Due to the increased maintence & upkeep, landsize, and resource usage needed per pool.

Thus, even though these two neighborhoods don't have a spray park, one can assume they are of higher wealth due to them both having two pools which require more resources than a spray park. Making them have a higher wealth compared to other neighborhoods such as Allegheny Center.
