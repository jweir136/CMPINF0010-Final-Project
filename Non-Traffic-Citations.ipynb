# Crime Analysis -- Non Traffic Citations

Although serious and violent crime is a big factor in detemrining best neighborhoods, It's also good to consider the much more common type of crime that would be seen often: non-traffic citations. These are low level crimes that might not directly threaten oneself, but will definitley affect quality of living in certain neighborhoods.



```python
#import statements
import pandas as pd 
%matplotlib inline
#Displaying general format of data
data = pd.read_csv("CrimeData.csv")
data.head(5)

```

```python
#Seperate data into neighborhood, and amount of NTC per neighborhood
df = data["NEIGHBORHOOD"].value_counts()
pd.set_option('display.max_rows', None)
#Remove Invalid data
df1 = df.drop(["Unable To Retrieve Address"])
print(df1)
```

```python
#Graph said data in bar graph
graph = df1.plot.barh(figsize = (15, df.shape[0] *0.3))
#Set labels
graph.set_xlabel("Neighborhood")
graph.set_ylabel("Amount of Non-Traffic Citations")
graph.set_title("Comparison of Neighborhood to Amount of Non-Traffic Citations")
```

Here, we get a good visuaization of the neighborhoods that are more more skewed to have higher amounts of non-traffic citations. Here, we see South Side Flats has a exponentially larger amount of these than any other neighborhood. This is not unexpected, however. South Side Flats is a neighborhood with vibrant nightlife that is home to many bars, smoke shops, etc. which may increase the amount of things like public drunkeness, and underage drinking. Other neighborhoods, like the Central Business Distric and Central Oakland have high amounts of crime like underage sale of alcohol, due to the high college population there.
